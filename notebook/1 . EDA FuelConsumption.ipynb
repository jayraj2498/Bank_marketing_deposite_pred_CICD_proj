### 2) Data Collection
- Dataset Source - 

- age: Age of the individual.
- job: Job type of the individual.
- marital: Marital status of the individual.
- education: Education level of the individual.
- default: Whether the individual has credit in default or not.
- balance: Balance in the individual's account.
- housing: Whether the individual has a housing loan or not.
- loan: Whether the individual has a personal loan or not.
- contact: Type of communication contact.
- day: Day of the month.
- month: Month of the year.
- duration: Duration of the last contact in seconds.
- campaign: Number of contacts performed during this campaign for this individual.
- pdays: Number of days that passed by after the client was last contacted from a previous campaign.
- previous: Number of contacts performed before this campaign for this individual.
- poutcome: Outcome of the previous marketing campaign.
- deposit: Whether the client subscribed to a term deposit or not.

```python
# Distribution of numerical features
numerical_features = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
plt.figure(figsize=(12, 10))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

```python

# Distribution of categorical features
categorical_features = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
plt.figure(figsize=(15, 12))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(3, 3, i)
    sns.countplot(x=feature, data=df)
    plt.title(f'Distribution of {feature}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Distribution of the target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='deposit', data=df)
plt.title('Distribution of Deposit (Target Variable)')
plt.xlabel('Deposit')
plt.ylabel('Count')
plt.show()
```

```python
# Select only numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Correlation matrix for numerical columns only
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Define numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Define categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

# Display numerical columns
print("Numerical columns:")
print(numerical_columns)

# Display categorical columns
print("\nCategorical columns:")
print(categorical_columns)
```

```python
# Assuming df is your DataFrame and categorical_columns contains categorical column names

# Find unique categories in each categorical column
unique_categories = {}
for col in categorical_columns:
    unique_categories[col] = df[col].unique()

# Print unique categories for each categorical column
for col, categories in unique_categories.items():
    print(f"Unique categories in '{col}':")
    print(categories)
    print()
```
